```python
#helpful: http://nbviewer.ipython.org/github/jvns/pandas-cookbook/blob/master/cookbook/Chapter%209%20-%20Loading%20data%20from%20SQL%20databases.ipynb
import pandas as pd
import sqlite3
import xlrd as xl
```

First, we just need to select a sample file. I'm using xls purely because pandas can select different sheets from .xls files and this will be useful when creating multiple tables within a single SQLite database

```python
path = ('car_sales.xlsx')
```

```python
# if this .sqlite db doesn't already exists, this will create it
# if the .sqlite db *does* already exist, this establishes the desired connection
con = sqlite3.connect("sqlite_cars.sqlite")
```

```python
# this pulls out the names of the sheets in the workbook. We'll use these to name the different tables in the 
# SQLite database that we'll create
sheet_names = xl.open_workbook(path).sheet_names()
```

I created this loop so that this particular .xls file is totally interchangeable with any other file if we use something else we like more as a sample. 
- Basically, it makes the script work even if a different file with a different # of sheets were being used

```python
for i in range(len(sheet_names)):
    sheet_name = sheet_names[i]
    df = pd.read_excel(path, sheetname='{}'.format(sheet_name))
    con.execute("DROP TABLE IF EXISTS {}".format(sheet_name))
    pd.io.sql.to_sql(df, "{}".format(sheet_name), con, index=False)
```

```python
# now the spreadsheets are in tables in a mini database!
```

Finally, this function isn't really necessary. I just added it because I know I'm  going to be too lazy to want to write out  

for every query. This just allows you to read the database by entering the desired query and turns the results into a pandas dataframe

```python
def run(query):
    results = pd.read_sql("{}".format(query), con)
    return results
```

### Some important notes about SQL:

- Data requests in SQL are called "queries"

- In SQL, there are databases and multiple tables within those databases. If you're used to Excel, think of a SQL database as a single Excel file, and each table is a different sheet within the same file

- Just to reiterate: "table" = "spreadsheet"

- Unlike Python, capitalization DOES NOT matter in SQL. However, over capitalization is arbitrarily used to distinguis the formal components of a query

- Also unlike Python, indentations and line separations are arbitrary. I opt to use structure to organize my thoughts, but it's not necessary

- Below, SQL queries appear between the triple quotes ('''), the rest is just the python code set up to make the queries run

### How to read queries:

Queries are much more like normal English syntax than most languages out there. Reading them out loud can help with understanding their structure. 

The two most basic components of a query are nearly always "SELECT" and "FROM".  Think about going to the grocery store to look for milk. This could be the structure of your search:

In English: 

SELECT = "show me"

FROM = "which can be found in"

Let's say you decide you only want 2% milk. You can specify that using WHERE:

Again in English:

SELECT = "show me"

FROM = "which can be found in"

WHERE = "but only if"    
    

#### Although capitalization and indentation do not matter, ***the order of the clauses cannot change***. You don't *need* every clause in a query to make it work, but if you do use it, it has to appear in the right order.

### Simple queries to start:

First, we'll look at everything  - all rows and columns - from the sales_table, basically like looking at a simple excel spreadsheet.

Every query will have SELECT and FROM - all other components are optional ways to get more specific
Translation: "Show me everything - all columns and rows - from the sales table"

```python

run('''
    SELECT
        *
    FROM
        sales_table
    ''')
```

```python
# add "WHERE" to specifically look at data that matches a certain criteria
# Translation: "Show me all columns and rows from the sales table that involves cash transactions ONLY"
run('''
    SELECT
        *
    FROM
        sales_table
    WHERE
        payment_type = 'cash'
    ''')
```

```python
# change "=" to "!=" in the "WHERE" clause to specify that you DON'T want to see a particular factor
# Translation: "Show me all columns and rows from the sales table that did not involve cash transactions ONLY"
run('''
    SELECT
        *
    FROM
        sales_table
    WHERE
        payment_type != 'cash'
    ''')
```

```python
# add "ORDER BY" to see the most to sort the data by a particular column 
# by default, ORDER BY sorts numbers in ascending order
# English Translation: Show me all columns and rows for cash transactions, starting with the lowest sale price
run('''
    SELECT
        *
    FROM
        sales_table
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue
    ''')
```

```python
# add "DESC" after the column name in the ORDER BY clause if you'd rather see the largest numbers first
run('''
    SELECT
        *
    FROM
        sales_table
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    ''')
```

```python
# add "LIMIT" to see only the first few rows from the table
# Translation: "I want to see all columns for the five cash transactions with the highest sale price"
run('''
    SELECT
        *
    FROM
        sales_table
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    LIMIT 
        5
    ''')
```

But maybe you don't need all the data from these columns - maybe you only want to see the model_id and the sale_price for the 5 cash transactions with the highest sales price. So, instead of *, you specify columns:


```python
run('''
    SELECT
        model_id,
        revenue
    FROM
        sales_table
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    LIMIT 
        5
    ''')
```

The model_id isn't very helpful for analysis - the name would be better. However, each car's Model and Make information isn't located on the sales_table, but rather on the car_table:

 

```python
run('''
    SELECT
        *
    FROM
        car_table
    ''')
```

You can add data from two different tables using a "JOIN" in the FROM clause:

```python
run('''
    SELECT
        make,
        model,
        revenue
    FROM
        sales_table
        JOIN car_table 
            ON sales_table.model_id = car_table.model_id
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    LIMIT 
        5
    ''')
```

Notes on what's happening here:
- To JOIN two tables, find columns on each table that contain corresponding data
- In this case, the model_id column in the sales_table matched up to the model_id column in the car_table. Both have the ID numbers that the dealership has assigned to each model of the cars they are selling.
- Follow the format below, where a column is referred to by table_name.column_name: 

- Take care: corresponding columns from different tables don't necessarily always have identical *names*. So, if the model_id column in the car_table were named car_model_id instead, this would be the FROM clause:

Let's say you wanted to see the model_id in the table as well. If you just try adding "model_id" in the SELECT clause, suddenly you'll hit an error:

```python
run('''
    SELECT
        model_id,
        make,
        model,
        revenue
    FROM
        sales_table
        JOIN car_table 
            ON sales_table.model_id = car_table.model_id
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    LIMIT 
        10
    ''')
```

This is because you've introduced another table with a column named "model_id" - the query doesn't know if it should pick the data from the model_id column in the sales_table or the car_table.  So, you have to specify which table you want the query to use:

```python
run('''
    SELECT
        sales_table.model_id,
        make,
        model,
        revenue
    FROM
        sales_table
        JOIN car_table 
            ON sales_table.model_id = car_table.model_id
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    LIMIT 
        5
    ''')
```

Note: because the data in the sales_table.model_id column corresponds to the data in car_table.model_id, you can specify either one in the SELECT clause in this case. There are instances where this won't be the case, but that can be addressed later.

*** You only need to specify the column's table name in the SELECT clause when there are one or more columns of the same name (so if the joined tables both have a column named "model_id", for instance). Because there is only one column named "revenue" between the two tables joined here, you don't have to specify the table. ***

You can also give tables nicknames so that you don't have to write out the entire name of the table each time. You can use the nickname in the SELECT clause, even though you don't state the nicknae until the FROM clause. Just put the nickname immediately after in the FROM clause. One or two letters for a nickname is pretty common:

```python
run('''
    SELECT
        S.model_id,
        make,
        model,
        revenue
    FROM
        sales_table S
        JOIN car_table 
            ON S.model_id = car_table.model_id
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    LIMIT 
        5
    ''')
```

Now that we can see the Make and Model information for each car, let's throw in the sticker_price as well - also located on the car_table:

```python
run('''
    SELECT
        make,
        model,
        revenue, 
        sticker_price
    FROM
        sales_table S
        JOIN car_table C
            ON S.model_id = C.model_id
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    LIMIT 
        5
    ''')
```

Now we can see that each car sold for some amount less than the listing price, presumably because the customer negotiated a better price with the salesman. You can add a function in the SELECT clause to see the percent under the listing price:

```python
## hacky fix, need to figure out the deal with integers and pandas.to_sql on SQLite
run('''
    SELECT
        make,
        model,
        revenue, 
        sticker_price,
        revenue*100/sticker_price
    FROM
        sales_table S
        JOIN car_table C
            ON S.model_id = C.model_id
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    LIMIT 
        5
    ''')
```

But that column name looks like a mess, so you can rename it by adding "as [desired_column_name]" 

```python
run('''
    SELECT
        make,
        model,
        revenue, 
        sticker_price,
        revenue*100/sticker_price AS percent_of_sticker_price
    FROM
        sales_table S
        JOIN car_table C
            ON S.model_id = C.model_id
    WHERE
        payment_type = 'cash'
    ORDER BY
        revenue DESC
    LIMIT 
        5
    ''')
```

Let's say you want to see the average price that each model of car sold for compared to the list price for the entire month. This is where the GROUP BY clause helps. We'll also need to use the AVG( ) function in the SELECT clause:

```python
# I've elimated the constraints from the WHERE, ORDER BY and LIMIT clause so all rows from the sale_table are included in the average
run('''
    SELECT
        make,
        model,
        AVG(revenue) AS average_revenue, 
        sticker_price,
        revenue*100/sticker_price AS percent_of_sticker_price
    FROM
        sales_table S
        JOIN car_table C
            ON S.model_id = C.model_id
    GROUP BY
        model
    ''')
```

The Average_Revenue numbers are messy, you can round them with ROUND(). To round to the nearest cent: ROUND( ,2)

```python
run('''
    SELECT
        make,
        model,
        ROUND(AVG(revenue), 2) AS average_revenue, 
        sticker_price,
        revenue*100/sticker_price AS percent_of_sticker_price
    FROM
        sales_table S
        JOIN car_table C
            ON S.model_id = C.model_id
    GROUP BY
        model
    ''')
```

ORDER BY the percent_of_sticker_price to highlight which cars are selling for the furthest below their list price:

```python
run('''
    SELECT
        make,
        model,
        ROUND(AVG(revenue), 2) AS average_revenue, 
        sticker_price,
        revenue*100/sticker_price AS percent_of_sticker_price
    FROM
        sales_table S
        JOIN car_table C
            ON S.model_id = C.model_id
    GROUP BY
        model
    ORDER BY 
        percent_of_sticker_price
    ''')
```

One last thing - we'll add a count of how many sales were made of each car model. If we use "id" column in car_sales, we can count

```python
run('''
    SELECT
        make,
        model,
        count() as units_sold,
        ROUND(AVG(revenue), 2) AS average_revenue, 
        sticker_price,
        revenue*100/sticker_price AS percent_of_sticker_price
    FROM
         sales_table S
            JOIN car_table C
                ON S.model_id = C.model_id 
    GROUP BY
        model
    ORDER BY 
        percent_of_sticker_price
    ''')
```

So for some reason, customers are negotiating harder (or more successfully) on the Toyota Camry and Tundra - and it would probably be causing a problem for the dealership. This kind of issue comes up a lot in B2B as well, since companies have bargaining power with their partners. 

Being able to pull this information quickly makes it easy to pinpoint problems, although more queries are then needed to dig into the reasons. The data in this SQLite db could help answer questions like:
- What age group is buying Tundras and Camrys? Maybe if younger people are drawn to them, they bargain harder because they have less expenadable income
- What salesperson has been selling these cars? Maybe he's not a very good negotiator

```python
run('''
    SELECT
        first_name,
        last_name,
        count() as units_sold,
        ROUND(AVG(revenue), 2) AS avg_revenue, 
        ROUND(AVG(sticker_price), 2) avg_sticker_price,
        ROUND(AVG(revenue)/AVG(sticker_price), 2) AS percent_of_sticker_price
    FROM
         sales_table S
            JOIN car_table C
                ON S.model_id = C.model_id 
            JOIN salesman_table SM 
                ON S.salesman_id = SM.id
    GROUP BY
        first_name
    ORDER BY 
        percent_of_sticker_price
    ''')
```

Although Claudine Hatch has good volume, she is giving up too much in negotations - she might need extra training. Meanwhile, Jared Case might have earned himself a bonus for performance this month.

What impact is Claudine's weak negotiating having on the overall margins for the dealership? How much could they gain if she got up to Jared's standards?  We'll include cogs to get some Net info:

```python
run('''
    SELECT
        first_name,
        last_name,
        count() as units_sold,
        sum(revenue) AS total_revenue, 
        sum(cogs) total_cogs, 
        sum(revenue) - sum(cogs) AS Net,
        (sum(revenue) - sum(cogs))*100 / sum(revenue) Margin,
        (sum(revenue)*100 / (select sum(revenue) from sales_table)) as percent_of_gross,
        (sum(revenue) - sum(cogs))*100 / (select sum(revenue)-sum(cogs) from sales_table S join car_table C on S.model_id = C.model_id) as percent_of_net
    FROM
         sales_table S
            JOIN car_table C
                ON S.model_id = C.model_id 
            JOIN salesman_table SM 
                ON S.salesman_id = SM.id
    GROUP BY
        first_name
    ORDER BY 
        total_revenue desc
    ''')
```

Claudine's sales make up 12% of gross revenues, but only contribute 10% to net - she's hurting margins and could benefit from training, but doesn't appear to be a lost cause.  However, Rosemarie Self has weak margins *and* is barely moving inventory. This might not be the right role for her. 

You could even use this data to figure out which salesperson would be best to approach a new customer who just walked in the door.
