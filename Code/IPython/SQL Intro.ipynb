## About SQL:

- Stands for "Structured Query Language"

- Can be pronounced "S. Q. L." or "sequel" 

- One of the most widely-used languages for managing relational databases

- A relational database has multiple tables. It's sort of like an Excel file:
    - Database = a single Excel file
    - Table = a single sheet in the same Excel file

- SQL is the language, but the software optimized for storing relational databases that you can access with SQL varies (MySQL, Microsoft SQL Server, Oracle, SQLite)
    - We're using SQLite in these examples

- The four basic functions you can perform with SQL are referred to as C.R.U.D.:
    - Create
    - Read
    - Update
    - Delete

- Requests are often called "queries"

- For a little more background, this is a quick but helpful read: http://sql.learncodethehardway.org/book/introduction.html

**STRUCTURE & FORMATTING:**

- INDENTATIONS and RETURNS:
    - indentations and returns are arbitrary in MySQL
    - a query can be written entirely on a single line or on multiple lines
    - most people opt to use structure to organize their thoughts (but not necessary)

- CAPITALIZATION:
    - Important:
        - Variables (anything that appears in quotation marks)
        - Nicknames
    - NOT Important:
        - Clauses like SELECT/select
        - Functions like SUM/sum
        - Column names like COLUMN_A/column_a
        - Table names like TABLE_A/table_a
- DATA TYPES:
    - Varies depending on the database software that you're using (MySQL, Microsoft, etc) - easy to google it 
    - SQLite (used in examples below):
        - To see the data types for each column in a table, run:
            - PRAGMA TABLE_INFO(sales_table)
        - Common data types:
            - TEXT: various characters. Dates will be saved as text in SQLite (major SQLite detractor)
            - INTEGER: integer
            - REAL: floating-point value
        - For more info: https://www.sqlite.org/datatype3.html

- ORDER OF COMMANDS:
    - Very strict
    - Not all clauses need to be present in a query, but when they are present, then must be in the following order:

### Setting up the SQLite Database:

It's not necessary to learn what's going on here (for now), just run the code to set up the database

```python
# Set up a SQLite database using several .csv files:

# if this .sqlite db doesn't already exists, this will create it
# if the .sqlite db *does* already exist, this establishes the desired connection
con = sqlite3.connect("sqlite_cars.sqlite")

# create dataframes from each .csv file:
sales_table = pd.read_csv('https://raw.githubusercontent.com/DaveBackus/Data_Bootcamp/master/Code/SQL/sales_table.csv')
car_table = pd.read_csv('https://raw.githubusercontent.com/DaveBackus/Data_Bootcamp/master/Code/SQL/car_table.csv')
salesman_table = pd.read_csv('https://raw.githubusercontent.com/DaveBackus/Data_Bootcamp/master/Code/SQL/salesman_table.csv')
cust_table = pd.read_csv('https://raw.githubusercontent.com/DaveBackus/Data_Bootcamp/master/Code/SQL/cust_table.csv')

# make a list of the tables (dataframes) and table names:
tables = [sales_table, car_table, salesman_table, cust_table]
table_names = ['sales_table', 'car_table', 'salesman_table', 'cust_table']

# drop each table name if it already exists to avoid error if you rerun this bit of code
# then add it back (or for the first time, if the table didn't already exist)
for i in range(len(tables)):
    table_name = table_names[i]
    table = tables[i]
    con.execute("DROP TABLE IF EXISTS {}".format(table_name))
    pd.io.sql.to_sql(table, "{}".format(table_name), con, index=False)

# Finally, a little function to make it easy to run queries on this mini-database
def run(query):
    results = pd.read_sql("{}".format(query), con)
    return results
```

##### "HAVING" in GROUP BY statements:

    SELECT
        column_a,
        SUM(column_b) 
    FROM
        table_name
    GROUP BY
        column_a HAVING SUM(column_b) > x   # only includes groups in column_a when the sum of column_b is greater than x 

If you've applied a function to a column and want to filter to only show results meeting a particular criteria, you will use GROUP BY instead of WHERE:

```python
run('''
    SELECT
        model_id, 
        sum(revenue)
    FROM
        sales_table
    GROUP BY
        model_id HAVING sum(revenue) > 300000
    ''')
# show me the total revenue per model_id from the sales_table
```

#### Rewrite the query above to look at average revenue per model_id, and filer it to model_ids whose average revenue is greater than 20,000:

```python
run('''
    SELECT
        model_id, 
        avg(revenue)
    FROM
        sales_table
    GROUP BY
        model_id HAVING avg(revenue) > 20000
    ''')
# show me the total revenue per model_id from the sales_table
```

##### Why use HAVING instead of WHERE?

If the column has a function applied to it, use "GROUP BY column HAVING...". If there's it no function, use WHERE. 

This is because of the order in which SQL pulls and manipulates the data...and it's not super intuitive. These are the steps SQL goes through with the following function:

    SELECT 
        model_id,            
        AVG(revenue)
    FROM
        sales_table       
    WHERE revenue > 20000
    GROUP BY
        model_id

1. Find the sales_table
2. Pull the data from the model_id and revenue columns
3. Filter out all rows where revenue isn't greater than 20000
4. Average the remaining rows by group

Even though AVG( ) appears pretty early in the script, it's not actually being applied until **after** the WHERE statement has filtered out the rows

Instead, use "GROUP BY column HAVING..." and SQL will follow these steps and give you what you're looking for:

    SELECT
        model_id, 
        avg(revenue)
    FROM
        sales_table
    GROUP BY
        model_id HAVING avg(revenue) > 20000

1. Find the sales_table
2. Pull the data from the model_id and revenue_columns
3. Average the rows by group (model_id)
4. Return rows where the model_id's average revenue is greater than 20,000

### ALIASES (or nicknames)

    SELECT
        column_a AS alias_a    # creates a nickname for column_a
    FROM
        table_name AS T        # creates a nickname for column_b
    WHERE
        alias_a = x            # now you can refer to column_a as alias_a in the rest of the query!!!

Aliases are always optional, but they can save you time and make the column headers to you result_set look a lot cleaner. 

http://www.w3schools.com/sql/sql_alias.asp

```python
run('''
    SELECT
        model_id,
        avg(revenue) AS rev_avg
    FROM 
        sales_table
    GROUP BY 
        model_id
    ''')
# Show me the average revenue per model_id, label the column head "rev_avg"
```

Once you've assigned an alias, you can use the alias later in the query. Below "rev_avg" appears again in the GROUP BY statement (instead of avg(revenue) )

```python
run('''
    SELECT
        model_id, 
        ROUND(AVG(revenue), 2) AS rev_avg
    FROM
        sales_table
    GROUP BY
        model_id HAVING rev_avg > 20000
    ''')
```

#### Rewrite the query above, with the following changes:

- group by date instead of model_id
- sum revenue instead of averaging it, then give it an alias (and use the alias for the rest of the query)
- limit the result-set to only include dates that have more than 80,000 in revenue
- order your results by so that the date with the most revenue is listed first

```python
run('''
    SELECT
        date, 
        sum(revenue) AS total_rev
    FROM
        sales_table
    GROUP BY
        date HAVING total_rev > 80000
    ORDER BY
        total_rev desc
    ''')
```
