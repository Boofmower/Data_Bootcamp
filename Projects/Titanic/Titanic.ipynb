## 2. Analyze and prepare data

### Clean up the data

Import the relevant libraries.

```python
import pandas as pd
import numpy as np
from sklearn import metrics
import os
```

Check the current working directory.

```python
os.getcwd()
```

Load the training dataset and view a sample.

```python
training = pd.read_csv('Data/train.csv', sep=',')
training.head(10)
```

The objective is to predict 'Survived' using the other variables. Looking at the above we can probably immediately drop the following:

* 'Name' - Doesn't make sense to dummy as each name is essentially unique. Also hard to imagine name alone correlates with survival rate after controlling for other proxies of social/economic status ('Pclass' and 'Fare')
* 'Ticket' - Doesn't seem useful at all...
* 'Cabin' - Doesn't seem particularly useful plus most are blank

Take a quick look at 'Cabin' to double-check.

```python
training['Cabin'].fillna(0).value_counts()
```

Drop the unnecessary columns before proceeding.

```python
training = training.drop(['Name', 'Ticket', 'Cabin'], axis=1)
training.head(10)
```

Now take a look to see if there is any missing data.

```python
training.info()
```

It appears that 'Age' and 'Embarked' have missing values. <b>Handling missing data could be an exercise by itself</b> (and we should do analysis to see if the missing rows might have a pattern) but let's just set missing quantitative values to 0 and set missing 'Embarked' values to S because it is the most frequent value.

```python
training['Embarked'].value_counts().plot(kind='bar')
```

```python
training['Embarked'] = training['Embarked'].fillna('S')
training = training.fillna(0)
training.info()
```

The algorithms require numerical inputs so we'll need to create dummy/indicator variables for 'Sex' and 'Embarked'.

```python
embarked_dummies = pd.get_dummies(training['Embarked'])
embarked_dummies.head(10)
```

Join back to main dataset and make sure the dummies joined correctly by looking at the 'Embarked' value and the associated 1/0's.

```python
training = training.join(embarked_dummies)
training.head(10)
```

Do the same for 'Sex' and drop the text columns.

```python
sex_dummies = pd.get_dummies(training['Sex'])
training = training.join(sex_dummies)
training = training.drop(['Embarked', 'Sex'], axis=1)
training.head(10)
```

ALMOST done! You need to delete one column per categorical variable due to the 'dummy variable trap'. If we were to run a regression with the above dataset we would have perfect multicollinearity in some cases. This is when independent variables are highly correlated which leads to bad things... More info without a bunch of math: http://www.investopedia.com/terms/m/multicollinearity.asp

For example, 'female' and 'male' now have a -1.0 correlation because each person is tagged as only either female or male.

```python
training = training.drop(['S', 'male'], axis=1)
training.head(10)
```

## 3. Evaluate algorithms

Let's approach this as a classification problem because we'll want to predict a binary response (survived or not). For this example we'll use a general purpose ensemble algorithm (random forest).

Before we begin let's split the dataset into data (independent variables) and target (dependent variable). For scikit-learn you'll need to convert from Pandas dataframes to arrays.

```python
training_data = training.values[:,2:]
training_target = training.values[:,1]
training_target[:5]
```

Bring in the test dataset and apply the same basic transformations.

```python
test = pd.read_csv('Data/test.csv', sep=',')
test = test.drop(['Name', 'Ticket', 'Cabin'], axis=1)
test['Embarked'] = test['Embarked'].fillna(method='ffill')
test = test.fillna(0)
embarked_dummies = pd.get_dummies(test['Embarked'])
test = test.join(embarked_dummies)
sex_dummies = pd.get_dummies(test['Sex'])
test = test.join(sex_dummies)
test = test.drop(['Embarked', 'Sex'], axis=1)
test = test.drop(['S', 'male'], axis=1)
test_data = test.values
test_data[:3]
```

### Random forest

An ensemble method which creates a number of decision trees on different subsets of the dataset. A fast, multi-purpose algorithm but generally a black-box. More info: http://en.wikipedia.org/wiki/Random_forest.

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators = 1000)
rf_model.fit(training_data, training_target)
print(rf_model)
```

Make predictions on the training dataset and summarize the fit of the model. Note there is probably some overfitting as we are still within the training dataset. Will apply the algorithm to the test set for final scoring on Kaggle. Normally you'd split your dataset of known values into train/test but skipping for this demonstration as going to show cross validation instead.

```python
expected = training_target
predicted = rf_model.predict(training_data)
print(metrics.accuracy_score(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
print(metrics.roc_auc_score(expected, predicted))
```

98% accuracy?!?! Is this for real?

A quick look at relative feature importance.

```python
print(rf_model.feature_importances_)
print(training.columns)
# Use np.savetxt to get header and combine so easier to read? Maybe zip the array?
```

Hard to read but seems that 'Age', 'Fare' and 'female' are the primary model drivers.

Because we modeled within-sample (didn't properly split train/test) it's hard to trust the accuracy. However, we can run cross validation as an estimate of model accuracy as if it were making predictions on previously unseen data. 79% accuracy seems much more realistic.

```python
from sklearn import cross_validation
from sklearn.linear_model import LogisticRegression

num_folds = 10
num_instances = len(training_data)
kfold = cross_validation.KFold(n=num_instances, n_folds=num_folds)
model = LogisticRegression()
results = cross_validation.cross_val_score(model, training_data, training_target, cv=kfold)
print(results)
print(results.mean())
print(results.std())
```

Use the algorithm on the true test data and combine/export for upload to Kaggle.

```python
predicted = rf_model.predict(test_data[:,1:])
submission = np.c_[test_data[:,0].astype(int), predicted.astype(int)]
submission = pd.DataFrame(submission, columns=['PassengerId', 'Survived'])
submission.to_csv('Data/submission_rf.csv', index=False)
submission.head()
```

So how'd we do? Overall a 74% accuracy at Kaggle on true test data. <u>Not very good</u> relative to some simple heuristics:

* A survival table created based on the proportion of survivors using gender, their ticket class and the price they paid for their ticket (78% accurate)
* Assume that if the passenger is female then survives, if not then does not (77% accurate)
* Assume that all did not survive (62% accurate; our baseline)

Remember how important it is to understand your data?
